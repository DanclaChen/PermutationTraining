```python
import numpy as np
np.set_printoptions(threshold=np.inf)

import matplotlib as mpl
import matplotlib.pyplot as plt
from palettable.cartocolors.sequential import DarkMint_4

mpl.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['mathtext.fontset'] = 'cm'

A = np.loadtxt('plot640.txt')
```

```python
A_sum = np.sum(A, axis=1)
print(A_sum)
```

```python
ax = plt.figure(figsize=(2, 6))
plt.gca().set_aspect(14)
plt.barh(range(len(A_sum[0:400])), A_sum[0:400], height=1, color='#137a63')
plt.xlabel('Frequency', size=20)
plt.xticks(fontsize=15)
plt.gca().set_ylim(0,400)
plt.yticks([])
plt.gca().invert_yaxis()
plt.yticks(fontsize=15)
plt.title('n = 640', fontsize=20)

plt.gca().set_aspect('auto')
plt.savefig('distribution640short.pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(16, 6), gridspec_kw={'width_ratios': [2, 4, 8, 2], 'wspace': 0.1})

A0 = np.loadtxt('plot160.txt')

axs[0].set_title('n = 160', fontsize=20)
axs[0].set_xlabel('Weights', fontsize=20)
axs[0].set_ylabel('Iteration of permutation', fontsize=20)
labels_pos = np.linspace(0, np.shape(A0)[1]-1, 3)
axs[0].set_xticks(labels_pos, ['-1', '0', '1'], fontsize=15)
axs[0].set_yticks([0, 100, 200, 300, 399], [0, 100, 200, 300, 400], fontsize=15)

axs[0].imshow(A0, cmap=DarkMint_4.mpl_colormap, aspect='auto')


A1 = np.loadtxt('plot320.txt')

axs[1].set_title('n = 320', fontsize=20)
axs[1].set_xlabel('Weights', fontsize=20)
labels_pos = np.linspace(0, np.shape(A1)[1]-1, 3)
axs[1].set_xticks(labels_pos, ['-1', '0', '1'], fontsize=15)
axs[1].set_yticks([])

axs[1].imshow(A1, cmap=DarkMint_4.mpl_colormap, aspect='auto')


A2 = np.loadtxt('plot640.txt')

axs[2].set_title('n = 640', fontsize=20)
axs[2].set_xlabel('Weights', fontsize=20)
labels_pos = np.linspace(0, np.shape(A2)[1]-1, 5)
axs[2].set_xticks(labels_pos, ['-1', '-0.5', '0', '0.5', '1'], fontsize=15)
axs[2].set_yticks([])

axs[2].imshow(A2, cmap=DarkMint_4.mpl_colormap, aspect='auto')


axs[3].barh(range(len(A_sum[0:400])), A_sum[0:400], height=1, color='#137a63')
axs[3].set_xlabel('Frequency', size=20)
axs[3].set_xticks([0, 1000], [0, 1000], fontsize=15)
plt.gca().set_ylim(0,400)
plt.yticks([])
plt.gca().invert_yaxis()
plt.yticks(fontsize=15)
plt.title('n = 640', fontsize=20)

plt.gca().set_aspect('auto')

axs[3].axhline(y=6, color='#cf4037', linewidth = 1.5, linestyle='--')
axs[3].axhline(y=50, color='#cf4037', linewidth = 1.5, linestyle='--')
axs[3].axhline(y=300, color='#cf4037', linewidth = 1.5, linestyle='--')

plt.savefig('IsPerm.pdf', bbox_inches='tight')
```
